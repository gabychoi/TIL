### apply 변환

- 행이나 열 단위로 복잡한 데이터 가공이 필요한 경우 사용하는 함수
- lambda 식
- apply 함수는 인자로 함수를 넘겨 받을 수 있다.

```python
def get_square(a) :
    return a**2
```

```python
print("제곱근 : ", get_square(3))
```

```python
# 위 코드를 람다식으로 바꿔 보자.
lambda_square = lambda a : a**2
print("제곱근 : " , lambda_square(3))
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age) :
    category = ""
    if age <= 5 :
        category = 'baby'
    elif age <= 12 :
        category = 'child'
    elif age <= 19 :
        category = 'teenager'
    elif age <= 24 :
        category = 'student'
    elif age <= 39 :
        category = 'young adult'
    elif age <= 60 :
        category = 'adult'
    else :
        category = 'elderly'
    return category
```

```python
titanic_reset_index_df['age_category'] = titanic_reset_index_df['age'].apply(lambda x : get_category(x))
titanic_reset_index_df['age_category'].value_counts
```

```python
# 승객의 나이와 성별에 따른 카테고리를 cat 으로 추가하라
# 조건 1) 20살 이상이면 성별을 그대로 사용
# 조건 2) 20살 미만이면 성별에 관계 없이 'child' 라고 정의하라

func1 = lambda df : df.sex if df.age >= 20 else 'child'
func2 = lambda df : "NaN" if pd.isnull(df.age) else func1(df) 

titanic['cat'] = titanic.apply(func2, axis=1)
titanic['cat'].value_counts()
```
