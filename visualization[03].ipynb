```python
# Feature Selection
# 학교명, 학교급구분, 소재지도로명, 위도, 경도

feature_sel_df = seoul_school_df[['학교명', '학교급구분', '소재지도로명주소', '위도', '경도']]
feature_sel_df.head()
```

```python
# 결측값 확인 및 제거
feature_sel_df.dropna(inplace=True)
print( feature_sel_df.isnull().sum() )
```

```python
# 소재지도로명주소가 서울을 포함하고 학교급구분 초등학교만 추출

seoul = feature_sel_df[feature_sel_df['소재지도로명주소'].str.contains('서울특별시')]
seoul_elem = seoul[seoul['학교급구분'] == '초등학교']
seoul_elem.head()
```

```python
# query 문 사용 (에러 해결 요망)
condition = " 학교급구분 == '초등학교' and 소재지도로명주소.str.contains('서울특별시').values"
result_df = feature_sel_df.query(condition)
result_df.head()
```
