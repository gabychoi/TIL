```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
# 타이타닉 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이 값으로 대체

titanic['age'] = titanic['age'].fillna(titanic['age'].mean()).astype(int)
titanic.head()
```

```python
# 타이타닉 승객에 대해 나이와 성별에 의한 age_gender_cat 열을 생성하라
# 조건 1) 성별을 나타내는 문자열 male 또는 female 로 시작한다
# 조건 2) 성별을 나타내는 문자열 뒤에 나이를 나타내는 문자열로 변경하라
# 조건 3) 예시) 남성의 나이가 27이라면 male27

titanic['age_gender_cat'] = titanic['sex'] + titanic['age'].astype(int).apply(str)
display(titanic['age_gender_cat'].head(10))

# 내 답안
titanic['age_str'] = titanic['age'].astype(int).apply(str)
titanic['age_gender_cat2'] = titanic['sex'] + titanic['age_str']

titanic['age_gender_cat2'].head()
```

### 데이터프레임 인덱스 조작하는 방법
- set_index : 기존 행 인덱스를 제거하고 데이터 열 중 하나를 인덱스로 설정
- reset_index : 기존 행 인덱스를 제거하고 인덱스를 데이터 열 추가

```python
# 1. 5명의 학생의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 만들어라.
# 1) 학생 이름을 나타내는 열을 포함시키지 않고 데이터프레임 df_score1 을 생성한 후, 
# 2) df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다. 
# 3) reset_index 명령으로 이 인덱스 열을 일반 데이터열로 바꾸여 데이터프레임 df_score2을 만든다.

np.random.randint(1, 101)
df_score1 = pd.DataFrame(np.round(np.random.rand(3,5), 2).T,
                       columns=['국어', '영어', '수학'],
                        index=['호진', '재원', '선림', '한준', '선영'])

df_score1.index.name = '이름'
df_score2 = df_score1.reset_index()
df_score2
```

```python
# 2. 학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 df_score2에 
# set_index 명령을 적용하여 다시 학생 이름을 나타내는 열을 인덱스로 변경한다.

df_score3 = df_score2.set_index('이름')
display(df_score3)
```

### DataFrame merge

```python
data1 = {
    '학번' : [1,2,3,4],
    '이름' : ['섭섭해', '김한준', '김선림', '최호진'],
    '학년' : [2,4,1,3]
}

data2 = {
    '학번' : [1,2,4,5],
    '학과' : ['CS', 'MATH', 'MATH', 'CS'],
    '학점' : [2.4, 4.5, 3.4, 4.2]
}
```

```python
stu_df   = pd.DataFrame(data1)
major_df = pd.DataFrame(data2)
display(stu_df)
display(major_df)
```

```python
# 특정 컬럼 기준으로 매칭되는 것만 merge가 된다.
pd.merge(stu_df, major_df)
```

```python
# merge 시 on, how 옵션을 줄 수 있다.
pd.merge(stu_df, major_df, on='학번', how='inner') # inner join : 일치하는 데이터만 merge
```

```python
pd.merge(stu_df, major_df, on='학번', how='left') # left join : 왼쪽에 있는 stu_df 기준으로 merge
```

```python
pd.merge(stu_df, major_df, on='학번', how='right') # right join : 오른쪽에 있는 major_df 기준으로 merge
```

```python
pd.merge(stu_df, major_df, on='학번', how='outer') # outer join : 전체 데이터 merge
```

- 컬럼 인덱스가 아닌 인덱스를 기준으로 병합하라.
- left_index , right_index

```python
pop_df01 = pd.DataFrame({
    'city' : ["seoul", 'seoul','seoul', 'busan', 'busan'], 
    'year' : [2010, 2005, 2020, 2018, 2015],
    'pop' : [1234567, 2345678, 356789, 4567890, 5678901]
})
pop_df01
```

```python
pop_df02    = pd.DataFrame( np.arange(12).reshape((6,2)) , 
    index   = [['busan' , 'busan', 'seoul' , 'seoul' , 'seoul' , 'seoul'] , 
               [2010, 2005, 2020, 2018, 2015, 2010]] , 
    columns =  ['col01','col02'])
pop_df02
```

```python
pd.merge(pop_df01, pop_df02, left_on=['city', 'year'], right_index=True) 
```

- 다중 인덱스가 아닌 단일 인덱스를 이용하는 방법

```python
data1 = { "이름" : ["이지안","박동훈","이순신","강감찬"],
          "학년" : [2,4,1,3]}

data2 = { "학과" : ["CS","MATH","MATH","CS"],
          "학점" : [3.4,2.9,4.5,1.2]}
```

```python
df01 = pd.DataFrame(data1, index=[1,2,3,4])
df02 = pd.DataFrame(data2, index=[1,2,4,5])
```

```python
display(df01)
display(df02)
```

```python
merge_df = pd.merge(df01, df02, right_index=True, left_index=True)
merge_df
```

```python
# join 함수 이용 가능 , default 가 outer join 이다
df01.join(df02 , how='inner')
```
