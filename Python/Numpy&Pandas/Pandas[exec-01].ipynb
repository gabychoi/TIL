```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 각 파일을 pd.read_csv() 로 읽어들인다
ratings_df = pd.read_csv('C:/Users/i/data/ratings.dat', sep='::', header=None)
movies_df = pd.read_csv('C:/Users/i/data/movies.dat', sep='::', header=None)
users_df = pd.read_csv('C:/Users/i/data/users.dat', sep='::', header=None)
```

```python
del users_df[3]
del users_df[4]
# OR
# users_df.drop([3,4], axis=1, inplace=True)
```

```python
# 데이터 확인 및 구조분석과 각 데이터 프레임에 헤더를 추가 ratings.dat
# userId / movieId / rating / timestamp
ratings_df.columns = ['userId', 'movieId', 'rating', 'timestamp']
movies_df.columns = ['movieId', 'title', 'genres']
users_df.columns = ['userId', 'gender', 'age']
```

```python
# 공통 컬럼을 이용한 데이터 병합
merge_df = pd.merge(ratings_df, movies_df, on = 'movieId', how='outer') 
merge_df = pd.merge(merge_df, users_df, on = 'userId', how='outer') 

#or
merge_df = pd.merge(pd.merge(ratings_df, users_df), movies_df)
merge_df.head()
```

```python
# user 평점 갯수 순위 확인
user_sort = pd.DataFrame(ratings_df.groupby('userId').size())
user_sort.head()
user_sort.sort_values(by=0 , ascending=False).head()
```

```python
# 각 영화의 성별 평균 평점

rating_mean_df = merge_df.pivot_table('rating', 'title', 'gender', aggfunc=np.mean)
rating_mean_df
```

```python
# 평점이 220개 이상인 영화만 필터링

print( merge_df.groupby('title').size())
print( merge_df.groupby('title').size().index )
rating_by_title = merge_df.groupby('title').size()
rating220 = rating_by_title.index[rating_by_title >= 220]
```

```python
# 평점이 200개 이상인 영화들의 성별 평균 평점
rating200 = rating_by_title.index[rating_by_title >= 200]
rating200

rating200_mean = rating_mean_df.loc[rating200]
rating200_mean.head()
```

```python
# 여성이 가장 좋아하는 영화 순위
rating_mean_df.sort_values(by=['F'], ascending=False)
```

```python
# 남녀간의 평균 평점 차이
rating_mean_df['diff_rating'] = np.abs(rating_mean_df['F'] - rating_mean_df['M'])
rating_mean_df['diff_rating'].head()
```
