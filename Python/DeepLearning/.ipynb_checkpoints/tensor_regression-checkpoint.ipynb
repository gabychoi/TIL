```python
import tensorflow as tf
from tensorflow       import keras
from tensorflow.keras import layers

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
print(tf.__version__)
```

- Auto MPG Dataset

```python
# 파일 다운
dataset_path = keras.utils.get_file("auto-mpg.data", "http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data")
dataset_path
```

```python
columns = ['mpg' , 'cylinder' , 'displacement' , 'horsepower' , 'weight' , 'acceleration' , 'model year' , 'origin']

row_dataset = pd.read_csv(dataset_path, names=columns, sep=" ", comment='\t', skipinitialspace=True) #skipinitialspace=True 로 앞단의 공백 제거
row_dataset.head()
```

```python
dataset = row_dataset.copy()
dataset.tail()
```

```python
# 데이터 정제
# 결측값 확인
dataset['horsepower'].replace('?', np.nan , inplace=True)
```

```python
dataset.dropna(subset=['horsepower'], axis=0, inplace=True)
```

```python
dataset['horsepower'] = dataset['horsepower'].astype('float')
```

```python
dataset.info()
```

```python
dataset['horsepower'].unique()
```

```python
dataset.isna().sum()
```

```python
# origin 은 수치형이 아닌 범주형이므로 원-핫-인코딩을 통한 변환 필요
# 1 == USA , 2 == Europe , 3 == Korea
origin = dataset.pop('origin') # 컬럼 삭제
dataset['usa'] = (origin==1) * 1.0
dataset['europe'] = (origin==2) * 2.0
dataset['korea'] = (origin==3) * 3.0
```

```python
dataset.tail()
```

```python
# 세트 분할, tensorflow 와 sklearn 을 같이 사용할 수 X
train_set = dataset.sample(frac=.8 , random_state=0)
test_set  = dataset.drop(train_set.index)
```

```python
sns.pairplot(train_set[['mpg' , 'cylinder' , 'displacement' , 'weight']])
```

```python
sns.pairplot(train_set[['mpg' , 'cylinder' , 'displacement' , 'weight']] , diag_kind='kde')
```

```python
train_stats = train_set.describe()
train_stats.pop('mpg')
train_stats = train_stats.transpose() #행과 열 전치
```

```python
y_train = train_set.pop('mpg')
y_test = test_set.pop('mpg')
y_train.head()
y_test.head()
```

```python
# 데이터 정규화
def norm(x) :
    return (x - train_stats['mean']) / train_stats['std'] #정규화식
norm_train_set = norm(train_set)
norm_test_set = norm(test_set)
```

```python
norm_train_set.pop('mpg')
norm_test_set.pop('mpg')
```

```python
# 모델 생성
model = keras.Sequential([
    layers.Dense(64 , activation='relu' , input_shape=[len(norm_train_set.keys())]), # 활성화 함수
    layers.Dense(64 , activation='relu'),
    layers.Dense(1)
])

optimizer = tf.keras.optimizers.RMSprop() #최적화 함수

# 학습
model.compile(loss='mse',
              optimizer = optimizer ,
              metrics = ['mae' , 'mse'])
```

```python
model.summary()
```

```python
# 모델 학습
class PrintDot(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs):
    if epoch % 100 == 0: print('')
    print('.', end='')
    
# 훈련 정확도와 검증 정확도 history
history = model.fit(norm_train_set , y_train , epochs=500 , validation_split=.2 , verbose=1)
```

```python
pd.DataFrame(history.history).tail(10)
```

```python
sample_result = model.predict(norm_train_set[:10])
sample_result
```

```python
# 예측
loss , mae , mse = model.evaluate(norm_test_set , y_test , verbose=0)
print('평균 절대 오차 : ' , mae)
```

```python
import matplotlib.pyplot as plt

# 시각화
y_pred = model.predict(norm_test_set).flatten()
# y_pred
plt.scatter(y_test , y_pred)
plt.xlim([0, plt.xlim()[1]])
plt.ylim([0, plt.ylim()[1]])
plt.scatter(y_test , y_pred)
_ = plt.plot([-100 , 100] , [-100 , 100])
plt.show()
```
