```python
tmp_raw = []
for file in files :
    tmp = pd.read_excel(file, header=2)
    tmp_raw.append(tmp)
tmp_raw
```

```python
oil_station = pd.concat(tmp_raw)
oil_station.head()
```

```python
oil_station.info()
```

```python
#휘발유만 대상으로 분석 진행한다면?
stations = pd.DataFrame({'oil_store' : oil_station['상호'],
                                         'address' : oil_station['주소'],
                                         'price' : oil_station['휘발유'],
                                         'self_y_n' : oil_station['셀프여부'],
                                         'company' : oil_station['상표']})
stations.head()
```

```python
# 파생변수 생성
# 구별 주유 가격 조사를 위한 
stations['direct'] = [ addr.split()[1] for addr in stations['address']]
stations.head()
```

```python
# 구별 정보 확인
stations['direct'].unique()
```

```python
# 서울특별시 -> 성동구
# 특별시 -> 도봉구 로 변경하라

stations.loc[stations['direct']=='서울특별시'  , 'direct'] = '성동구'
stations.loc[stations['direct']=='특별시'  , 'direct'] = '도봉구'
```

```python
# 구별 정보 확인
stations['direct'].unique()
```

```python
stations[stations['price'] == '-']
```

```python
# 가격 정보가 기입되지 않은 주유소는 제거
stations = stations[stations['price'] != '-']
```

```python
# 컬럼 타입을 숫자형으로 변경
stations['price'] = [int(price) for price in stations['price']]
```

```python
# index 처리
stations.reset_index(inplace=True)
# stations.drop('index', axis=1)
del stations['index']
```

```python
stations.head()
```

```python
stations.boxplot(column='price' , by='self_y_n')
plt.show()
```

```python
# seaborn boxplot() 구현
plt.figure(figsize=(15,5))
sns.boxplot(x='company' , y='price' , data=stations , palette='Set3')
sns.swarmplot(x='company' , y='price' , data=stations , palette='Set3' , color='.6')
plt.show()
```
