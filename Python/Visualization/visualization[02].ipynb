- pie chart : 카테고리별 값을 상대적 비교할 때 활용
- 차트의 원형을 유지하려면 plt.axis('equal') 옵션을 준다.

```python
labels = ['뽕사부', '순대국밥', '개구리비빔밥', '생고기', '지하철버거']
sizes = [15, 15, 20, 25, 25]
colors = ['gold', 'blue', 'red', 'yellow', 'green']
plt.title('Pie Chart')
plt.pie(sizes, labels = labels, colors=colors, shadow=True, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

- histogram : 구간에 대한 데이터 집계를 시각화할 때 활용

```python
data = np.random.randn(1000)
data
plt.title('histogram')
plt.hist(data, bins=20)
plt.grid()
plt.show()
```

- scatter plot : 2차원 데이터 (X, Y)
- 데이터 집합의 상관관계를 살펴보기 위한 그래프

```python
np.random.seed(100)
X = np.random.normal(0,1,100)
Y = np.random.normal(0,1,100)
plt.title('scotter plot')
plt.scatter(X, Y)
plt.show()
```

```python
np.random.seed(100)
X = np.random.rand(30)
Y1 = np.random.rand(30)
Y2 = np.random.rand(30)
Y3 = np.pi * (15 * np.random.rand(30))**2
plt.title('bubble chart')
plt.scatter(X, Y1, c=Y2, s=Y3)
plt.show()
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

```python
iris.sepal_length[:20].plot(kind='bar', rot=0)
plt.show()
```

```python
iris[:5].plot.bar(rot=0)
# or 
iris[:5].plot(kind='bar', rot=0)
plt.show()
```

```python
# 그룹 분석으로 각 붓꽃 종의 평균을 구하고 시각화하라.
```

```python
iris_bar_df = iris.groupby('species').mean()
iris_bar_df.columns.name = 'feature'
iris_bar_df.plot.bar(rot=0)
plt.title('붓꽃 종별 평균')
plt.show()
```

```python
iris_bar_df = iris.groupby('species').mean()
iris_bar_df.columns.name = 'feature'
iris_bar_df.plot.barh(rot=0)
plt.title('붓꽃 종별 평균')
plt.show()
```

```python
iris_bar_df.T.plot.bar(rot=0) # 전치행렬로 데이터 바꾸기
plt.show()
```

- pie, hist, box, scatter

```python
titanic.head()
# 선실별 승객 수 비율을 파이 차트로 구현해보자.
pclass_ratio_cnt = titanic.pclass.value_counts()
# type(pclass_ratio_cnt)
pclass_ratio_cnt.plot.pie(autopct='%.2f%%',
                          startangle=45,
                          legend=True,
                          shadow=True,
                          labels=pclass_ratio_cnt.index,
                          colors=['skyblue', 'pink', 'green'])
plt.axis('equal')
plt.title('title')
plt.show()

# 파이 차트는 비율을 볼 때, 시리즈 형식의 데이터를 넣어 쓴다. NO DataFrame
```

#### autopct 사용법
- '%0.2f' — 소숫점 두번째자리 까지 보여준다. ex ) 4.08
- '%0.2f%%' — 소숫점 두번째자리까지 보여주고 맨 뒤에 ‘%’ 기호를 붙인다. ex ) 4.08%
- (% 기호를 두 번 쓰는 이유는 문자열 포매팅 할 때 처음 % 기호가 이스케이프 문자처럼 작동하기 때문이다.)
- '%d%%' — 가장 가까운 정수로 반올림하고 맨 뒤에 % 기호를 붙인다. ex) 4%
