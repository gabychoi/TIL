```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 키(meter)와 몸무게(kg)
heights = [[1.6] , [1.65] , [1.7] , [1.77] , [1.85]]
weights = [[60] , [65] , [70] , [77] , [85]]
```

```python
plt.title('키 & 몸무게 그래프')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.plot(heights, weights, 'k.')
plt.axis([1.5, 1.90, 55, 90])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression() # 모델 생성
# fit 함수는 리스트 또는 배열 형태의 인수를 필요로 한다.
lr_model.fit(heights , weights) # fit(독립변수, 종속변수) 학습
```

```python
weight_pred = lr_model.predict([[1.7], [1.75]]) #예측
weight_pred # 예측값 출력
```

```python
plt.title('키 & 몸무게 그래프')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.grid(True)
plt.axis([1.5, 1.90, 55, 90])

plt.plot(heights, weights, 'k.')

# 선형 회귀 선을 그린다
plt.plot(heights , lr_model.predict(heights), color='blue')

plt.show()
```

### auto-mpg.csv 를 이용한 선형회귀

```python
import pandas as pd
import numpy as np

row_data = pd.read_csv('C:/Users/i/data/auto-mpg.csv' , header=None)
row_data.columns = ['mpg' , 'cylinder' , 'displacement' , 'horsepower' , 'weight' , 'acceleration' , 'model year' , 'origin' , 'name']
row_data.head()
```

```python
# 데이터의 자료형 확인
row_data.info()
```

```python
# 다른 독립변수들을 활용해 mpg(연비)를 예측 
```

```python
row_data['horsepower'].unique()
```

-  horsepower 컬럼에 대한 전처리
- 1) '?' -> np.nan 으로 변경
- 2) 행 삭제 
- 3) 문자열을 실수형으로 변환

```python
row_data['horsepower'].replace('?', np.nan , inplace=True)
```

```python
row_data['horsepower'].unique()
```

```python
row_data.dropna(subset=['horsepower'], axis=0, inplace=True) #subset으로 'horsepower' 만 특정해 dropna 해줌 
```

```python
row_data['horsepower'] = row_data['horsepower'].astype('float')
```

```python
row_data.info()
```

```python
# 분석에 활용할 변수를 선택
cor_df = row_data[['mpg' , 'cylinder' , 'horsepower' , 'weight']]
cor_df.head()
```

```python
# 선형 관계 파악을 위한 산점도 그래프 생성
cor_df.plot(kind='scatter' , x='weight' , y='mpg' , c='red' , s=10 , figsize=(10, 5))
plt.show()
```

```python
cor_df.plot(kind='scatter' , x='cylinder' , y='mpg' , c='red' , s=10 , figsize=(10, 5))
plt.show()
```

```python
cor_df.plot(kind='scatter' , x='horsepower' , y='mpg' , c='red' , s=10 , figsize=(10, 5))
plt.show()
```

```python
# seaborn 산점도
fig = plt.figure(figsize=(10,5))
area01 = fig.add_subplot(1, 2, 1)
area02 = fig.add_subplot(1, 2, 2)
sns.regplot(x='weight' , y='mpg' , data=cor_df , ax=area01) # 회귀선 포함 산점도
sns.regplot(x='weight' , y='mpg' , data=cor_df , ax=area02 , fit_reg=False) # fit_reg=False 로 회귀선 제거
plt.show()
```

```python
sns.pairplot(cor_df) # 모든 경우의 수를 그래프로 생성
plt.show()
```

```python
# X = cor_df['weight'] # 단일선형회귀
X = cor_df[['cylinder', 'horsepower', 'weight']] # 여러 독립변수 / 다중선형회귀 
y = cor_df['mpg'] #종속변수
```

```python
# 데이터 세트를 구분
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X, y, test_size=0.2, random_state=10)

print(len(X_train))
print(len(X_test))
print(X_train)
print(X_test)
```

```python
from sklearn.linear_model import LinearRegression
auto_lr_model = LinearRegression()
```

```python
auto_lr_model.fit(X_train , y_train)
r_square = auto_lr_model.score(X_test , y_test)
print('결정계수 : ' , r_square)
```

```python
# 회귀의 기울기
print('기울기 : ' , auto_lr_model.coef_)
# 회귀의 절편
print('절편 : ' , auto_lr_model.intercept_)
```

```python
# 모델에 전체 X 데이터를 입력해 예측값 , 실제 y 
y_pred = auto_lr_model.predict(X)
```

```python
result_df = pd.DataFrame({
    'guess' : y_pred,
    'answer' : y
})
```

```python
sns.distplot(y , hist=False , label='answer')
sns.distplot(y_pred , hist=False , label='guess')

plt.show()
```
