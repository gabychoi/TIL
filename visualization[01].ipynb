- option(color , marker , style)
- color : 약어 -> blue : b , red : r , green : g 라고 약어를 준다.
- marker : o , s(square) 등 
- style : line style 의미 , - , -- , : 
- lw : 선 굵기
- ls : 선 스타일
- ms : 마커 크기
- mec : 마커 선 색깔
- mcw : 마커 선 굵기
- mfc : 마커 내부 색깔

```python
plt.title('라인 플롯')
plt.plot([10,20,30,40] , [1,4,9,16] , 'bs--') #앞이 x축, 뒤가 y축, 선 스타일 지정
plt.xlabel('x 라벨')
plt.ylabel('y 라벨')
plt.show()
```

```python
plt.title('스타일 적용 예')
plt.plot([10,20,30,40] , [1,4,9,16] , 
         c = 'b' , lw = 5 , ls = '--' , marker = 'o' , ms = 15 , mec = 'g' , mew = 5 , mfc = 'r') #앞이 x축, 뒤가 y축, 선 스타일 지정
plt.xlabel('x 라벨')
plt.ylabel('y 라벨')
plt.xlim()
plt.ylim()
plt.show()
```

- 여러개의 라인 플롯을 그려보자.

```python
data = np.arange(0.0, 5.0 , 0.2)
data
plt.title('여러개의 라인 플롯')
plt.plot(data , data, 'r--', data, data**2, 'b--', data, data**3, 'g--')
plt.show()
```

```python
plt.title('라인을 겹쳐 그리기')
plt.plot([1,4,9,16] , 
         c = 'b' , lw = 5 , ls = '--' , marker = 'o' , ms = 15 , mec = 'g' , mew = 5 , mfc = 'r') 
plt.plot([9,16,4,1] , 
         c = 'k' , lw = 3 , ls = '--' , marker = 's' , ms = 10 , mec = 'm' , mew = 5 , mfc = 'c')
plt.show()
```

```python
X = np.linspace(-np.pi , np.pi , 256)
Y_C , Y_S = np.cos(X) , np.sin(X)
plt.plot(X , Y_C , ls='--' , label='cosine')
plt.plot(X , Y_S , ls='--' , label='sine')
plt.legend(loc=0) # loc로 범례의 위치 지정
plt.show()
```

- subplot(2,1,1)
- subplot(2,1,2)

```python
X1 = np.linspace(0.0, 5.0)
Y1 = np.cos(2 * np.pi * X1) * np.exp(-X1)
print(X1)
print(Y1)
X2 = np.linspace(0.0, 2.0)
Y2 = np.cos(2 * np.pi * X2)
print(X2)
print(Y2)
```

```python
axes01 = plt.subplot(1,2,1) # 그래프를 그리는 캔버스 생성
plt.plot(X1, Y1)
plt.title('subplot 01')

axes02 = plt.subplot(1,2,2)
plt.plot(X2, Y2)
plt.title('subplot 02')

plt.tight_layout() # 자동으로 간격을 맞춰주는 기능
plt.show()
```

```python
# bar plot
# x 데이터는 카테고리 값인 경우가 대부분이다.

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pylab as plt
%matplotlib inline

Y = [2, 3, 1]
X = np.arange(len(Y))
X

xlabel = ['1등실', '2등실', '3등실']
plt.title('bar plot')
plt.bar(X, Y) 
plt.xticks(X, xlabel)
plt.yticks(sorted(Y))
plt.xlabel('선실 등급')
plt.ylabel('생존자 수')
plt.show()
```

```python
# barh plot : horizental 형식으로 그래프를 그려준다.

Y = [2, 3, 1]
X = np.arange(len(Y))
X

xlabel = ['1등실', '2등실', '3등실']
plt.title('bar plot')
plt.barh(X, Y) 
plt.xticks(X, xlabel)
plt.yticks(sorted(Y))
plt.xlabel('선실 등급')
plt.ylabel('생존자 수')
plt.show()
```
